```python
def get_category_dictionaries(url, attribute_class):

    html = requests.get(url)
    soup = BeautifulSoup(html.text, "html.parser")
    table = soup.find('table', attrs={"class":attribute_class})
    category_dict = {}
    
    for row in table.findAll('tr'):
        columns = row.findAll('td')
        if (columns):
            category_link = columns[0].find('a')
            category = str(category_link['title'])
            links = columns[1].findAll('a');   
            link_dict = {}
            for link in links:
                link_dict[str(link['title']) ]= str(link['href'])
            category_dict[category] = link_dict
            
    return category_dict
```

```python
def filter_dict(dict, require=None, exclude=None):
    dict_copy = copy.deepcopy(dict)
    
    for key, value in dict.iteritems():
        if require and require not in key:
            del dict_copy[key]
        if exclude and exclude in key:
            del dict_copy[key]
    
    return dict_copy
```

```python
def df_replace_column_name_substring(df, regex, replace):
    column_name_list = df.columns.tolist()
    #print len(column_name_list)
        
    new_column_names = []
    for column_name in column_name_list:
        new_column_name = re.sub(regex, replace, column_name)
        #print new_column_name
        new_column_names.append(new_column_name)
        #print new_column_names
        
        #print len(new_column_names)
    df.columns = new_column_names
    return df
```
